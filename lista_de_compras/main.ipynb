```python
from adicionar_item import adicionar_item
from listar_itens import listar_itens
from marcar_como_comprado import marcar_como_comprado
from remover_item import remover_item
```

```python
lista_compras = []
```

```python
def main():
    while True:
        print("\nMenu:")
        print("1. Adicionar item")
        print("2. Listar itens")
        print("3. Marcar item como comprado")
        print("4. Remover item")
        print("5. Sair")
        opc = input("Escolha: ")

        match opc:
            case "1":
                adicionar_item(lista_compras)
            case "2":
                listar_itens(lista_compras)
            case "3":
                marcar_como_comprado(lista_compras)
            case "4":
                remover_item(lista_compras)
            case "5":
                break
            case _:
                print("Opção inválida. Tente novamente.")
```

```python
if __name__ == "__main__":
    main()
```
